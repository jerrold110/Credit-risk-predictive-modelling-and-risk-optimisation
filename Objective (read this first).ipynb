### Project objective
The purpose of this predictive modelling project is simply not to predict if a client will default on their loan by determining their credit class (good or bad).

Rather, the purpose of this project is to maximize profits for a hypothetical bank gained through interest from clients who take a loan and succesfully repay them. This hypothetical bank has a limit on the amount of capital that can be used to give credit and has to minimise the amount of risk while maximising potential profit.

### Variables for prediction
The variables of interest are credit_amount and class. By creating a data driven model that can predict the probability of a client successfully repaying a loan, we can selectively create the best balance of profit to risk so as to maximise profit to the business while mitigating capital loss risk.

Clients who have a higher probability of defaulting will have to pay higher interest rates, these clients represent greater sources of value to the business, up to a certain level of risk, where the probability of default on a loan is too risky to give credit.

### Interest rate charges
To simulate realistic bank charges, credit interest rates are a function of credit amount, duration, and probability of payment (POD) and duration rise. The interest exponentially increases as default risk (POD) rises **which is in line with real banking practices**.
The interest charges will be:
* Base of 2% up to a value of 10 time units, and an additional 0.5% for every additional part of 10 time units
* No additional interest for POD values below and including 0.1, additional $ e^(POD-0.1)) $% for POD values above 0.1

A loan of 20,000 for 11 time units at 0.2 POD has to pay: $ (2+0.5)+[e^(0.2-0.1)]=3.61 $ %

### Risk factor and profit calculation
Risk is a function of amount of capital the bank risks losing and the probability of default. A client with 0% POD represents no risk, and any interest gained from this loan is, riskless (free) profit.
Risk factor = probability of default * amount of credit

Profit calculation = capital gained from interest

Based on these formulas for risk factor calculation, the risk accrued by the bank in the following two scenario for credit given by the theoretical bank are identical:

* 2 loans for 10,000 with a POD of 0.1 at 10 time units. 
* Interest = 2(10000*0.02) = 400
* 1 loan for 10,000 with a POD of 0.2 at 10 time units. 
* Interest = 10000*3.61 = 310

The interest gained from the first scenario (400) is more than in the second scenario (310), so the first scenario is more profitable to the bank
