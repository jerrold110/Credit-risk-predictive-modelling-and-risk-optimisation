```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('credit.csv')
```

```python
string_cols = ['credit_history', 'purpose', 'own_telephone',
       'personal_status', 'other_parties', 
       'other_payment_plans', 'housing', 'job','foreign_worker']

num_cols = [x for x in list(data.columns) if x not in string_cols]
```

```python
credit = data[num_cols].copy()
```

```python
credit.columns
```

```python
# changing appropriate categorical columns to numerical 
rep = {
    'checking_status':{'\'no checking\'':0, '\'>=200\'':2, '\'0<=X<200\'':1  ,'\'<0\'':-1},
    'savings_status':{'\'no known savings\'':0, '\'<100\'':50 ,'\'100<=X<500\'':350, '\'500<=X<1000\'':750, '\'>=1000\'':1500},
    'employment':{'unemployed':0, '\'<1\'':0.5, '\'1<=X<4\'':2.5, '\'4<=X<7\'':5.5, '\'>=7\'':8.5},
    'property_magnitude': {'\'real estate\'':3,'car':2,'\'life insurance\'':1,'\'no known property\'':0}
}

credit.replace(rep, inplace=True)
```

```python
# Converting BAD to 1 GOOD to 0
credit.replace({'class':{'good':0, 'bad':1}}, inplace=True)
```

```python
# Create a no checking status column
credit['no_checking'] = credit['checking_status'].apply(lambda x: 1 if x==0 else 0)
```

```python
# Create an average 'installment_commitment' by personal_status feature
credit['commit_by_p_status'] = data[['personal_status', 'installment_commitment']].groupby(
    'personal_status').transform(np.mean)
```

```python
numeric = credit
numeric.to_csv('data/data_numeric.csv', index=False)
```
